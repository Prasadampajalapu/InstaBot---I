## Answer 6.1

### Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
#going back to homepage
driver.get("https://www.instagram.com/")
```

```python
# Define a funnction to print the 1st 500 followers names

def get_username(profile):
    search_and_open(profile) #function defined above
    
    #Locate the Followers tab and click on it
    wait = WebDriverWait(driver, 15)
    path='/html/body/div[1]/section/main/div/header/section/ul/li[2]'
    followers=wait.until(EC.presence_of_element_located((By.XPATH,path)))
    followers.click()
    
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'isgrP')))
    fullbody=driver.find_element(By.CLASS_NAME,'isgrP')#full body of the followers by using class
    user_list=[]
    while len(user_list)<500:
        # By using BeautifulSoup,get the classes of the every follower
        data=driver.page_source 
        html_data=BeautifulSoup(data,'html.parser')
        ul = html_data.find_all(class_ = 'd7ByH')

        for u in ul[len(user_list):]:
            if u.a.string not in user_list: 
                user_list.append(u.a.string)

        #scrolling
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', fullbody)
    return user_list
```

```python
users = get_username('foodtalkindia')

print('The usernames of the first 500 followers of foodtalkindia are:')
for i in range(500):
    print(i+1, users[i])
```

## Answer 6.2

### Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
#going back to homepage
driver.get("https://www.instagram.com/")
driver.maximize_window()
```

```python
# Define a function for followers who don't followme
def not_follow_back(profile):
    search_and_open(profile) #search_and_open function already defined above
    
    #locate the followers whom i follow and open
    wait=WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'tc8A9'))).click()
    
    #Locate and click see all followers
    path='/html/body/div[6]/div/div/div[2]/div/a'
    wait.until(EC.presence_of_element_located((By.XPATH,path))).click()
    
    #to get all the followers whom i follow and they dont follow mw back
    followers_list=[]
    for i in range(1,101):#for sample take 100
        path1='/html/body/div[6]/div/div/div[2]/ul/div/li['+str(i)+']/div/div[2]/button'
        wait.until(EC.presence_of_element_located((By.XPATH,path1)))
        followers=driver.find_element(By.XPATH,path1)
        if followers.text=='Follow':
            break
        followers_list.append(driver.find_element(By.XPATH,'/html/body/div[6]/div/div/div[2]/ul/div/li['+str(i)+']/div/div[1]/div[2]/div[1]/span/a').get_attribute('title'))
    return followers_list
    
```

```python
users=not_follow_back('foodtalkindia')
print('All the followers of “foodtalkindia” that I am following but those who don’t follow me are:')
for i in users:
    print(i)
```

```python
from selenium.common.exceptions import TimeoutException
#Define a function to get story
def get_story(profile):
    search_and_open(profile) #function defined above
    wait = WebDriverWait(driver, 10)
    time.sleep(5)
    try:
        #finding element for the story
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'h5uC0')))
        
         
        #height and width different for the story viewed and not viewed
        if driver.find_element(By.CLASS_NAME,'CfWVH').get_attribute('height')=='291':
            print("You have already seen the story")

        else :
            print("You have not seen the story")
            story.click()
            print("You are now seening the story")

    except TimeoutException:
        print("The user has no story")
```
